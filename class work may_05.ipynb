```python
import pandas as pd
```

```python
data = pd.read_csv("Bankloan.txt")
data.head()
```

```python
data.isnull().sum()
```

```python
data[data.isnull().any(axis = 1)]
```

```python
data['age'] = data['age'].fillna(data['age'].mean())
data['ed'] = data['ed'].fillna(data['ed'].mean())
data['income'] = data['income'].fillna(data['income'].mean())
```

```python
data[data.isnull().any(axis = 1)]
```

```python
x = data.drop("default", axis = 1)
y = data.default
```

```python
y = pd.to_numeric(y, errors='coerce')
y = y.dropna()
x = x.loc[y.index]
```

### Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 3)
```

```python
logmodel = LogisticRegression()
logmodel.fit(xtrain, ytrain)
```

```python
ypred = logmodel.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(ytest, ypred)
print("Accuracy :", accuracy)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(xtrain, ytrain)
```

```python
ypred = decision_tree.predict(xtest)
ypred_train = decision_tree.predict(xtrain)
```

```python
accuracy = accuracy_score(ytest, ypred)
print("Test Accuracy :", accuracy)
accuracy = accuracy_score(ytrain, ypred_train)      # Overfitting
print("Train Accuracy :", accuracy)
```

```python
decision_tree = DecisionTreeClassifier(max_depth = 5)
decision_tree.fit(xtrain, ytrain)
```

```python
ypred = decision_tree.predict(xtest)
ypred_train = decision_tree.predict(xtrain)
accuracy = accuracy_score(ytest, ypred)
print("Test Accuracy :", accuracy)
accuracy = accuracy_score(ytrain, ypred_train)
print("Train Accuracy :", accuracy)
```

### Neural Network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(xtrain)
x_test_scaled = scaler.transform(xtest)
```

```python
model = Sequential([
    Dense(32, input_dim = 8, activation = 'relu'),
    Dense(16, activation = 'relu'),
    Dense(8, activation = 'relu'),
    Dense(4, activation = 'relu'),
    Dense(1, activation = "sigmoid")
])
```

```python
model.compile(optimizer = 'sgd', loss = 'binary_crossentropy', metrics =['accuracy'])
```

```python
trains = model.fit(x_train_scaled, ytrain, epochs = 80)
```

```python
test_loss = model.evaluate(x_test_scaled, ytest)
print(test_loss)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(trains.history['accuracy'])
plt.xlabel("Epochs")
plt.ylabel("accuracy")
plt.show()
```

```python
plt.plot(trains.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("loss")
plt.show()
```
